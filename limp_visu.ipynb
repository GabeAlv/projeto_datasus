### Pré-processamento e engenharia de atributos

```python
colunas_remover = ['ESC2010', 'ESCFALAGR1', 'CODMUNRES', 'LOCOCOR', 'NATURAL', 'CODMUNOCOR', 'OCUP']

dataset.drop(columns=colunas_remover, inplace=True, axis=1)
```

```python
dataset.head()
```

```python
dataset['DTOBITO'] = pd.to_datetime(dataset['DTOBITO'])

dataset['MES_OBITO'] = dataset['DTOBITO'].dt.month

dataset.fillna('desconhecido', inplace=True)
```

```python
dataset_2020 = dataset[dataset['ANO'] >= 2020]
```

```python
# 1. Defina as colunas a serem verificadas
colunas_para_limpar = ['SEXO', 'RACACOR', 'ESTCIV', 'ESC']

# 2. Defina os valores que, se encontrados em qualquer uma dessas colunas, devem levar à exclusão da linha
valores_a_excluir = ['desconhecido', 'Ignorado', 'Ignorada']

# 3. Cria a Máscara (Filtro)
# A função 'isna()' e 'apply(any)' é usada para criar uma Máscara Booleana
# onde TRUE significa que a linha DEVE ser removida.
mascara_remocao = dataset_2020[colunas_para_limpar].isin(valores_a_excluir).any(axis=1)

# 4. Aplica a Remoção: ~ inverte a máscara, mantendo as linhas 'FALSE' (limpas)
df_totalmente_limpo = dataset_2020[~mascara_remocao].copy() 

print(f"Linhas removidas: {mascara_remocao.sum()}")
```

### SALVANDO DATASET

```python
df_totalmente_limpo.to_csv('df_2020_limpo.csv')
```

```python
# GRÁFICO 1: Distribuição de Idade por Sexo
# Insight: Qual a faixa etária mais vulnerável em cada gênero?
import matplotlib.pyplot as plt
import seaborn as sns
# GRÁFICO: Distribuição lado a lado (mais fácil de comparar)
df_totalmente_limpo['ANO'] = pd.to_numeric(df_totalmente_limpo['ANO'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.histplot(data=df_totalmente_limpo, 
             x='IDADE', 
             hue='SEXO', 
             multiple="dodge",  # <- MUDANÇA AQUI: lado a lado
             bins=10, 
             palette="viridis")
plt.title('Distribuição de Casos por Faixa Etária e Sexo - 2020 em diante')
plt.xlabel('Idade')
plt.ylabel('Número de Casos')
plt.tight_layout()
plt.show()

# Confirmar os números
print("\nTotal por sexo (2020+):")
print(df_totalmente_limpo['SEXO'].value_counts())
```

```python
# GRÁFICO 2: Estado Civil (Vulnerabilidade Social)
# Insight: Pessoas solteiras ou casadas estão mais representadas?

plt.figure(figsize=(10, 6))
sns.countplot(data=df_totalmente_limpo, 
              y='ESTCIV', 
              order=df_totalmente_limpo['ESTCIV'].value_counts().index, 
              palette="rocket")
plt.title('Ocorrências por Estado Civil - 2020 em diante')
plt.xlabel('Contagem')
plt.ylabel('Estado Civil')
plt.tight_layout()
plt.show()

# Opcional: Mostra os valores exatos
print("\nDistribuição por Estado Civil (2020+):")
print(df_totalmente_limpo['ESTCIV'].value_counts())
```

```python
# GRÁFICO 3: Sazonalidade (Mês do Óbito)
# Insight: Existe algum mês com pico de casos? (Ex: início ou fim de ano)

plt.figure(figsize=(10, 6))
sns.countplot(data=df_totalmente_limpo, 
              x='MES_OBITO', 
              palette="Blues_d",
              order=sorted(df_totalmente_limpo['MES_OBITO'].dropna().unique()))
plt.title('Sazonalidade: Ocorrências por Mês - 2020 em diante')
plt.xlabel('Mês')
plt.ylabel('Número de Casos')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
top_causas = df_totalmente_limpo['CAUSABAS'].value_counts().nlargest(5).index
sns.countplot(data=df_totalmente_limpo[df_totalmente_limpo['CAUSABAS'].isin(top_causas)], 
              x='CAUSABAS', 
              order=top_causas, 
              palette="magma")
plt.title('Top 5 Códigos CID (Causa Básica) - 2020 em diante')
plt.xlabel('Código CID-10')
plt.ylabel('Frequência')
plt.xticks(rotation=45)  # Rotaciona os códigos pra ficar mais legível
plt.tight_layout()
plt.show()

# Opcional: Mostra os valores exatos
print("\nTop 5 causas de 2020 em diante:")
print(df_totalmente_limpo['CAUSABAS'].value_counts().nlargest(5))
```

```python
# 1. Passo Essencial: Filtrar as 5 causas mais comuns para que o gráfico não fique ilegível.
# GRÁFICO: Média de Idade por Método e Gênero (Top 5 Causas) - 2020 em diante

# Pega o top 5 causas DOS DADOS FILTRADOS
top_causas = df_totalmente_limpo['CAUSABAS'].value_counts().nlargest(5).index
df_top = df_totalmente_limpo[df_totalmente_limpo['CAUSABAS'].isin(top_causas)]

plt.figure(figsize=(10, 6))
sns.barplot(data=df_top, x='CAUSABAS', y='IDADE', errorbar=None, hue='SEXO', palette="Spectral") 

plt.title('Média de Idade por Método e Gênero (Top 5 Causas) - 2020 em diante', fontsize=14)
plt.xlabel('Código CID-10')
plt.ylabel('Idade Média')
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()

# Opcional: Mostra as médias exatas
print("\nMédia de idade por causa e sexo (2020+):")
print(df_top.groupby(['CAUSABAS', 'SEXO'])['IDADE'].mean().round(1))
```
